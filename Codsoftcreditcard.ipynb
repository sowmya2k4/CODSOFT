DROPPING THE ROWS THAT HAVE NAN VALUES

```python
df = data.dropna()
print(df)
```

```python
print("PIE CHART")
fig, ax = plt.subplots(1, 1)
ax.pie(df.Class.value_counts(),autopct='%1.1f%%', labels=['NORMAL','FRAUD'], colors=['yellowgreen','r'])
plt.axis('equal')
plt.ylabel('')
```

```python
fig, ax = plt.subplots(figsize = (20,20))
sns.heatmap(df.corr().round(0), cmap ="coolwarm" ,square= True, linewidth = 1 ,annot =  True)
```

**HIGH CORRELATION EXISTS BETWEEN THE VARIABLES**

```python
fig,(ax1,ax2) = plt.subplots(2,1, figsize = (10,15), sharex = True)
ax1.hist(df.Amount[df.Class == 0],bins = 50,color = 'y',alpha = 0.5)
ax1.set_yscale('log')
ax1.set_title('NORMAL')
ax1.set_ylabel('TRANSACTIONS')
ax2.hist(df.Amount[df.Class == 1],bins = 50,color  = 'b',alpha = 0.5)
ax2.set_yscale('log')
ax2.set_title('FRAUD')
ax2.set_xlabel('AMOUNT')
ax2.set_ylabel('TRANSACTIONS')
```

```python
df.drop(['Time'] ,axis = 1,inplace = True)
print(df)
```

**MODEL FITTING USING GAUSSIAN MODEL**

```python
 from sklearn.model_selection import train_test_split
 from sklearn.naive_bayes import GaussianNB
 from sklearn import metrics
```

```python
x = df.drop(['Class'],axis=1).values
y = df['Class'].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,random_state = 12)
```

```python
NB = GaussianNB()
NB.fit(x_train,y_train)
```

```python
y_pred = NB.predict(x_test)
y_pred
```

```python
print('ACCURACY',metrics.accuracy_score(y_test,y_pred) )
```

**THE ACCURACY IS 97% , THIS MODEL IS WELL FITTED FOR THE CLASSIFICATION**
